```python
import os
import numpy as np
import matplotlib.pyplot as pl
import seaborn as sns
from notebook_image_tabs import ImageTabs
```

```python
dump = 'dump'
if not os.path.exists(dump):
    os.makedirs(dump)

data2 = []
for p in range(1, 6):
    uniform_data = np.random.rand(10, 12)
    ax = sns.heatmap(uniform_data)
    ax.set_title('example #{:d}'.format(p))
    path = os.path.join(dump, 'heatmap_{:d}.png'.format(p))
    pl.savefig(path, bbox_inches='tight', pad_inches = 0.15)
    fig = pl.gcf()
    pl.close(fig)
    data2.append([str(p), path])
```

```python
dic_param2 = {
    'titleText': None,
    'borderPx': 1,
    'borderColor': 'gray',
    'tabBackgroundColor': 'white',
    'buttonMargin': 5,
    'buttonPaddingVert': 12,
    'buttonPaddingHori': 7,
    'buttonBorderColor': 'black',
    'imageWidthPerCent': 99,
    'width': 450,
    'height': 360,
    'selection': 4,
    
    'buttonColorBase': '#eeeeee',
    'buttonColorHover': '#dddddd',
    'buttonColorActive': '#bdbdbd',
}

t2 = ImageTabs(data=data2, params=dic_param2, iframe=False)
t2.show()
```
